```python
#Maths Scores Dataset - Important Features selection
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import ExtraTreesRegressor

# Build a forest and compute the feature importances
Maths_forest = ExtraTreesRegressor(n_estimators=10, criterion='mse', max_depth=None, min_samples_split=2, random_state=None)

Maths_features = NASdata_Maths_New1.drop('Maths %', axis = 1)
Maths_Scores = NASdata_Maths_New1['Maths %']

Maths_forest.fit(Maths_features, Maths_Scores)
importances = Maths_forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in Maths_forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(Maths_features.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
#From below Feature Ranking, we can say that the top 10 features which influences Maths scores are:
# Features: 0:State; 1:District; 20:Computer use; 21:Library use; 8:Father edu; 3:Age; 4:Category; 11:Mother occupation; 10:Father occupation; 9:Mother edu
```
